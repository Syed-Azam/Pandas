#### 8.2 Combining and Merging Datasets

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})
df2 = pd.DataFrame({'key': ['a', 'b', 'd'],
                    'data2': range(3)})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on='key')
```

```python
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})
display(df3)
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],
                    'data2': range(3)})
display(df4)
pd.merge(df3, df4, left_on='lkey', right_on='rkey')
```

```python
display(df1)
display(df2)
pd.merge(df1, df2, how='inner')
```

```python
# Table 8-1. Different join types with how argument
# Option Behavior
# 'inner' Use only the key combinations observed in both tables
# 'left' Use all key combinations found in the left table
# 'right' Use all key combinations found in the right table
# 'output' Use all key combinations observed in both tables together
```

#### Merging on Index

```python
left1 = pd.DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'],
                      'value': range(6)})
right1 = pd.DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
display(left1)
display(right1)
pd.merge(left1, right1, left_on='key', right_index=True)
```

```python
pd.merge(left1, right1, left_on='key', right_index=True, how='outer')
```

```python
# Using the indexes of both sides of the merge is also possible:
left2 = pd.DataFrame([[1., 2.], [3., 4.], [5., 6.]],
                      index=['a', 'c', 'e'],
                      columns=['Ohio', 'Nevada'])
right2 = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]],
                      index=['b', 'c', 'd', 'e'],
                      columns=['Missouri', 'Alabama'])
display(left2)
display(right2)
pd.merge(left2, right2, how='outer', left_index=True, right_index=True)
```

```python
left2.join(right2, how='outer')
```

```python
display(left1)
display(right1)
left1.join(right1, on='key')
```

```python
another = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [16., 17.]],
                       index=['a', 'c', 'e', 'f'],
                       columns=['New York', 'Oregon'])
another
```

```python
display(left2.join(right2))
left2.join([right2, another])
```

#### Concatenating Along an Axis

```python
s1 = pd.Series([0, 1], index=['a', 'b'])
s2 = pd.Series([2, 3, 4], index=['c', 'd', 'e'])
s3 = pd.Series([5, 6], index=['f', 'g'])
pd.concat([s1, s2, s3])
```

```python
pd.concat([s1, s2, s3], axis=1)
```

```python
s4 = pd.concat([s1, s3])
s4
```

```python
pd.concat([s1, s4], axis=1)
```

```python
pd.concat([s1, s4], axis=1, join='inner')
```

```python
pd.concat([s1, s4], axis=1, join_axes=[['a', 'b', 'f']])
```

```python
#  create a hierarchical index on the concatenation axis
result = pd.concat([s1, s1, s3], keys=['one', 'two', 'three'])
result
```

```python
result.unstack()
```

```python
pd.concat([s1, s2, s3], axis=1, keys=['one', 'two', 'three'])
```

```python
df1 = pd.DataFrame(np.arange(6).reshape(3, 2), index=['a', 'b', 'c'],
                  columns=['one', 'two'])
df2 = pd.DataFrame(5 + np.arange(4).reshape(2, 2), index=['a', 'c'],
                  columns=['three', 'four'])
display(df1)
display(df2)
pd.concat([df1, df2], axis=1, keys=['level1', 'level2'])
```

```python
#If you pass a dict of objects instead of a list, the dict’s keys will be used for the keys
pd.concat({'level1': df1, 'level2': df2}, axis=1, sort=True)
```

```python
pd.concat([df1, df2], axis=1, keys=['level1', 'level2'],
              names=['upper', 'lower'], sort=True)
```

```python
df1 = pd.DataFrame(np.random.randn(3, 4), columns=['a', 'b', 'c', 'd'])
df2 = pd.DataFrame(np.random.randn(2, 3), columns=['b', 'd', 'a'])
display(df1)
display(df2)
pd.concat([df1, df2], ignore_index=True, sort=True)
```

#### Combining Data with Overlap

```python
a = pd.Series([np.nan, 2.5, np.nan, 3.5, 4.5, np.nan],
              index=['f', 'e', 'd', 'c', 'b', 'a'])
b = pd.Series(np.arange(len(a), dtype=np.float64),
              index=['f', 'e', 'd', 'c', 'b', 'a'])
b[-1] = np.nan
display(a)
display(b)

np.where(pd.isnull(a), b, a)
```

```python
df1 = pd.DataFrame({'a': [1., np.nan, 5., np.nan],
                    'b': [np.nan, 2., np.nan, 6.],
                    'c': range(2, 18, 4)})
df2 = pd.DataFrame({'a': [5., 4., np.nan, 3., 7.],
                    'b': [np.nan, 3., 4., 6., 8.]})
display(df1)
display(df2)
df1.combine_first(df2)
```

### 8.3 Reshaping and Pivoting

```python
data = pd.DataFrame(np.arange(9).reshape((3, 3)),
                    index=pd.Index(['Ohio', 'Colorado', 'Karachi'], name='state'),
                    columns=pd.Index(['one', 'two', 'three'],
                    name='number'))
data
```

```python
# Using the stack method pivots the columns into the rows, producing a Series:
result = data.stack()
result
```

```python
result.unstack()
```

```python
result.unstack(0)
```

```python
s1 = pd.Series([0, 1, 2, 3], index=['a', 'b', 'c', 'd'])
s2 = pd.Series([4, 5, 6], index=['c', 'd', 'e'])
data2 = pd.concat([s1, s2], keys=['one', 'two'])
data2
```

```python
data2.unstack()
```

```python
data2.unstack().stack()
```

```python
data2.unstack().stack(dropna=False)
```

```python
df = pd.DataFrame({'left': result, 'right': result + 5},
                  columns=pd.Index(['left', 'right'], name='side'))
df
```

```python
df.unstack()
```

```python
df.unstack('state')
```

```python
df.unstack('state').stack('side')
```

### Pivoting “Long” to “Wide” Format

```python
data = pd.read_csv('macrodata.csv')
data
```

```python
periods = pd.PeriodIndex(year=data.year, quarter=data.quarter, name='date')
columns = pd.Index(['realgdp', 'infl', 'unemp'], name='item')
data = data.reindex(columns=columns)
data.index = periods.to_timestamp('D', 'end')
ldata = data.stack().reset_index().rename(columns={0: 'value'})
ldata[:10]
```

```python
pivoted = ldata.pivot('date', 'item', 'value')
pivoted
```

### Pivoting “Wide” to “Long” Format

```python
df = pd.DataFrame({'key': ['foo', 'bar', 'baz'],
                   'A': [1, 2, 3],
                   'B': [4, 5, 6],
                   'C': [7, 8, 9]})
df
```

```python
melted = pd.melt(df, ['key'])
melted
```

```python
reshaped = melted.pivot('key', 'variable', 'value')
reshaped
reshaped.reset_index()
```
